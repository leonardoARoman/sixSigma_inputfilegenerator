```python
# get new created csv file and work with it.

t_df = pd.read_csv('./transiant.csv')

racks = ['O13','O14','O15','O17','O18','O19','O21','O22',
         'R12','R13','R14','R16','R17','R18','R20','R21']

urls = ['kw_to_O13.csv','kw_to_O14.csv','kw_to_O15.csv',
        'kw_to_O17.csv','kw_to_O18.csv','kw_to_O19.csv',
        'kw_to_O21.csv','kw_to_O22.csv','kw_to_R12.csv',
        'kw_to_R13.csv','kw_to_R14.csv','kw_to_R16.csv',
        'kw_to_R17.csv','kw_to_R18.csv','kw_to_R20.csv',
        'kw_to_R21.csv']
```

```python
# Helper function: input file generator.

def generate_csv_files(rackList, urlList, kw):
    for i, item in enumerate(rackList):
        df = pd.read_csv('./transiant.csv')
        df.power[df.rack==rackList[i]] = kw
        df.to_csv('./generated_csv_files/'+kw+'_'+urlList[i])
```

```python
generate_csv_files(racks, urls, '12')
```
