```python
# Scatter plot helper function
us = []
def grapher(frame,rack):
    plannings = ['7','14','21','28','35','42']
    color = ['DarkGreen','DarkRed','Yellow','Pink','Black','Red']
    label = ['_u7','_u14','_u21','_u28','_u35','_u42']
    u = frame.plot.scatter(x='Time', y=rack+':Capacity Planning0 Temperature', color='DarkBlue', label=rack+'_u0');
    us.append(u)
    for i, item in enumerate(plannings):
        #df = pd.read_csv('./transiant.csv')
        u = frame.plot.scatter(x='Time', y=rack+':Capacity Planning'+plannings[i]+' Temperature', color=color[i], label=rack+label[i],ax = us[i]);
        us.append(u)
    return us[6]
```

```python
import plotly as py
import plotly.graph_objs as go

py.offline.init_notebook_mode(connected = True)

def heatMap(dataframe,n):
    trace = go.Heatmap(z=[df['O13:Capacity Planning'+n+' Temperature'],df['O14:Capacity Planning'+n+' Temperature'], 
                      df['O15:Capacity Planning'+n+' Temperature'],df['O17:Capacity Planning'+n+' Temperature'], 
                      df['O18:Capacity Planning'+n+' Temperature'],df['O19:Capacity Planning'+n+' Temperature'],
                      df['O21:Capacity Planning'+n+' Temperature'],df['O22:Capacity Planning'+n+' Temperature'], 
                      df['R12:Capacity Planning'+n+' Temperature'],df['R13:Capacity Planning'+n+' Temperature'],
                      df['R14:Capacity Planning'+n+' Temperature'],df['R16:Capacity Planning'+n+' Temperature'],
                      df['R17:Capacity Planning'+n+' Temperature'],df['R18:Capacity Planning'+n+' Temperature'],
                      df['R20:Capacity Planning'+n+' Temperature'],df['R21:Capacity Planning'+n+' Temperature']],
                   x=df['Time'],
                   y=['O13_u0','O14_u0','O15_u0','O17_u0','O18_u0','O19_u0','O20_u0','O21_u0',
                      'R12_u0','R13_u0','R14_u0','R16_u0','R17_u0','R18_u0','R20_u0','R21_u0'])
    data=[trace]
    return data
```

```python
# HEAT MAP FOR ALL RACKS TAKING U0 AS REFERENCE

py.offline.iplot(heatMap(df,'0'), filename='labelled-heatmap')
```
